```python
httomo_folder = None
edges = None
```

## Check Parameters

```python
assert httomo_folder, "Need to provide path to httomo folder"
assert edges is not None, "Need to provide string with comma-separaed edges, i.e. Pb-La,Br-Ka"
edge_list = edges.split(",")
```

```python
import h5py
import imageio
import pathlib
import numpy as np
import scipy.ndimage as ndi
import matplotlib.pyplot as plt
```

```python
volumes_list = {}
for edge in edge_list:
    httomo_outpath = f"{httomo_folder}/{edge}/{edge}.h5"
    with h5py.File(httomo_outpath, "r") as f:
        volumes_list[edge] = np.nan_to_num(f["data"][:])
```

```python
ncols = 2
nrows = len(volumes_list) // ncols + len(volumes_list) % ncols
```

## Transverse plane

```python
fig, axes = plt.subplots(ncols=ncols, nrows=nrows, figsize=(ncols*2, nrows*2), dpi=150)
if axes.ndim == 1:
    axes = np.array([axes])
for i in range(ncols*nrows):
    r,c = i//ncols, i%ncols
    if i >= len(volumes_list):
        axes[r,c].axis("off")
        break
    edge = edge_list[i]
    vol = volumes_list[edge]
    axes[r,c].set_title(edge)
    axes[r,c].imshow(vol[:,vol.shape[1]//2,:], cmap="gray", interpolation="none")
    axes[r,c].axis("off")
plt.savefig("./_assets/recons_transverse.png", bbox_inches="tight", dpi=300)
plt.show()
```

## Sagittal plane

```python
fig, axes = plt.subplots(ncols=ncols, nrows=nrows, figsize=(ncols*2, nrows*2), dpi=150)
if axes.ndim == 1:
    axes = np.array([axes])
for i in range(ncols*nrows):
    r,c = i//ncols, i%ncols
    if i >= len(volumes_list):
        axes[r,c].axis("off")
        break
    edge = edge_list[i]
    vol = volumes_list[edge]
    axes[r,c].set_title(edge)
    axes[r,c].imshow(vol[vol.shape[0]//2,:,:], cmap="gray", interpolation="none")
    axes[r,c].axis("off")
plt.savefig("./_assets/recons_sagittal.png", bbox_inches="tight", dpi=300)
plt.show()
```

## Coronal plane

```python
fig, axes = plt.subplots(ncols=ncols, nrows=nrows, figsize=(ncols*2, nrows*2), dpi=150)
if axes.ndim == 1:
    axes = np.array([axes])
for i in range(ncols*nrows):
    r,c = i//ncols, i%ncols
    if i >= len(volumes_list):
        axes[r,c].axis("off")
        break
    edge = edge_list[i]
    vol = volumes_list[edge]
    axes[r,c].set_title(edge)
    axes[r,c].imshow(vol[:,:,vol.shape[2]//2].T, cmap="gray", interpolation="none")
    axes[r,c].axis("off")
plt.savefig("./_assets/recons_coronal.png", bbox_inches="tight", dpi=300)
plt.show()
```
